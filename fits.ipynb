```python
from timer import Timer
import numpy as np
import pandas as pd

from highly_adaptive_regression import HighlyAdaptiveLassoCV
from kernel_ridge import HighlyAdaptiveRidgeCV, RadialBasisKernelRidgeCV, MixedSobolevRidgeCV 
from kernel_ridge import kernels

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.dummy import DummyRegressor

import pandas as pd
import glob
import os
import altair as alt
```

```python
LEARNERS = {
    # 'Mean':DummyRegressor(strategy="mean"),
    'Ridge Regression':Ridge(alpha=1e-3),
    'Random Forest':RandomForestRegressor(n_estimators=2000, n_jobs=-1),
    'Radial Basis KRR':RadialBasisKernelRidgeCV(
        gammas=[0.001, 0.01, 0.1, 1, 10], 
        eps=1e-10,
        max_alpha_coef_norm = 10,
    ),
    'Mixed Sobolev KRR':MixedSobolevRidgeCV(
        eps=1e-6,
        max_alpha_coef_norm = 100,
    ),
    'HAL':HighlyAdaptiveLassoCV(),
    'HAR':HighlyAdaptiveRidgeCV(eps=1e-10, order=0),
}
```

```python
def mu(X):
    return np.sin(2*np.pi*X) * (X>0) + (-X) * (X<=0)
n = 500
n_ = 500

results = []
X_ = np.linspace(-1,1,num=n_)
truth = mu(X_)

for rep in range(3):
    X = np.random.uniform(-1,1,size=(n,1))
    Y = mu(X[:,0]) + np.random.normal(scale=0.5, size=(n,))

    for name, learner in LEARNERS.items():
        learner.fit(X,Y)
        results.append(pd.DataFrame({
            'rep':rep,
            'learner':name,
            'x':X_,
            'truth':truth,
            'prediction':learner.predict(X_.reshape(-1,1)),
        }))

df = pd.concat(results, ignore_index=True)
```

```python
alt.data_transformers.disable_max_rows()

# Base chart for the predictions (learners) and the truth line
base = alt.Chart(df).encode(
    x='x',
    y=alt.Y('prediction', scale=alt.Scale(domain=[-1.25, 1.25], clamp=True), title='Prediction'),
    # color=alt.value("#a2d5d7"),
    color = alt.Color('rep:N', legend=None)
)

# Prediction line
pred_line = base.mark_line()

# Truth line (black)
truth_line = alt.Chart(df).mark_line(color='black',strokeWidth=3).encode(
    x='x',
    y=alt.Y('truth', title='Truth')
)

# Combine both the prediction and truth line before faceting
layered_chart = alt.layer(pred_line, truth_line)

# Apply faceting with wrapping using facet's 'columns' parameter
final_chart = layered_chart.facet(
    facet=alt.Facet('learner:N', title='Method'),
    columns=3,  # This will wrap the facets into 3 columns
    spacing=5
).resolve_scale(
    y='shared'  # Ensure y-scale is shared across facets
).resolve_scale(
    y='shared'
).configure(
    axis=alt.AxisConfig(labelFontSize=14, titleFontSize=16),     # Axis labels and titles
    title=alt.TitleConfig(fontSize=20),                          # Chart title
    legend=alt.LegendConfig(labelFontSize=14, titleFontSize=16)  # Legend labels and titles
).configure_header(
    labelFontSize=14,  # Facet label font size
    titleFontSize=16   # Facet title font size
)

final_chart
```
